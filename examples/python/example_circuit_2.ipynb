# Example 2 - Time Domain Steady State Analysis

```python
from CircuitCalculator.Circuit.circuit import Circuit
from CircuitCalculator.Circuit import components as ccp
from CircuitCalculator.SimpleAnalysis.TimeSeries import steady_state_voltage_timedomain_analysis, steady_state_current_timedomain_analysis, plot_timeseries_by_id
from CircuitCalculator.Circuit.solution import TimeDomainSolution
from CircuitCalculator.SignalProcessing.periodic_functions import CosFunction
import numpy as np
```

```python
circuit = Circuit([
    ccp.periodic_current_source(id='Iq', nodes=('0', '1'), wavetype=CosFunction.wavetype, I=1, w=5000, phi=0),
    ccp.resistor(id='R', R=2, nodes=('1', '0')),
    ccp.inductance(id='L', L=100e-6, nodes=('1', '0')),
    ccp.ground(nodes=('0',))
])
time_domain_solution = TimeDomainSolution(circuit=circuit, w_max=10000)
```

```python
_ = steady_state_current_timedomain_analysis(
    plot_timeseries_by_id('Iq'),
    plot_timeseries_by_id('L', color='green'),
    solution=time_domain_solution,
    tmax=0.005
)
```

```python
_ = steady_state_voltage_timedomain_analysis(
    plot_timeseries_by_id('L', color='red'),
    solution=time_domain_solution,
    tmax=0.005
)
```
