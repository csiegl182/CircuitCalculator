# Three-Phase Example (inductive loads in Y and Delta)

```python
import numpy as np

from CircuitCalculator.ThreePhaseCircuit import (
    ThreePhaseCircuit,
    three_phase_voltage_source_star,
    three_phase_impedance_load_star,
    three_phase_impedance_load_delta,
    three_phase_complex_solution,
)
```

```python
w = 2*np.pi*50
L_y = 0.03   # H
L_d = 0.02   # H

tp_circuit = ThreePhaseCircuit(
    components=[
        three_phase_voltage_source_star(id="V1", nodes=("BUS", "N"), V=230+0j, Z=0.2+0.05j),
        three_phase_impedance_load_star(id="Ly", nodes=("BUS", "N"), Z=1j*w*L_y),
        three_phase_impedance_load_delta(id="Ld", nodes=("BUS",), Z=1j*w*L_d),
    ],
    ground_node="N",
)

solution = three_phase_complex_solution(tp_circuit, w=w)
```

```python
for branch in ["Ly.a", "Ly.b", "Ly.c", "Ld.ab", "Ld.bc", "Ld.ca"]:
    I = solution.get_current(branch)
    print(branch, "|I|=", abs(I), "angle(deg)=", np.angle(I, deg=True))
```
