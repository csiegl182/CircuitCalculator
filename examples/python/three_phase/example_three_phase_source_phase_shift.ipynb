# Three-Phase Example (default source phase shift)

```python
from CircuitCalculator.ThreePhaseCircuit import (
    ThreePhaseCircuit,
    ThreePhaseVoltageSourceY,
    ThreePhaseVoltageSourceDelta,
    transform_three_phase_circuit,
)
```

```python
tp = ThreePhaseCircuit(
    components=[
        ThreePhaseVoltageSourceY(id="Vy", nodes=("BUSY", "N"), V=230+0j),
        ThreePhaseVoltageSourceDelta(id="Vd", nodes=("BUSD",), V=400+0j),
    ],
    ground_node="N",
)
c = transform_three_phase_circuit(tp)

for comp in c.components:
    if comp.type == "complex_voltage_source":
        v = complex(comp.value["V_real"], comp.value["V_imag"])
        print(f"{comp.id:6s} -> {v}")
```
