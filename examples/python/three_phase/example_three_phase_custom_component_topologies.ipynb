# Three-Phase Custom Component (line, star, delta)

```python
from CircuitCalculator.Circuit.Components import components as cp
from CircuitCalculator.ThreePhaseCircuit import (
    ThreePhaseCircuit,
    three_phase_custom_component_line,
    three_phase_custom_component_star,
    three_phase_custom_component_delta,
    transform_three_phase_circuit,
)
```

```python
line = three_phase_custom_component_line(
    id="L",
    nodes=("SRC", "LOAD"),
    phase_a=lambda cid, nd: cp.impedance(id=cid, nodes=nd, Z=0.1+0.2j),
    phase_b=lambda cid, nd: cp.impedance(id=cid, nodes=nd, Z=0.1+0.2j),
    phase_c=lambda cid, nd: cp.impedance(id=cid, nodes=nd, Z=0.1+0.2j),
)

star = three_phase_custom_component_star(
    id="S",
    nodes=("BUS", "N"),
    phase_a=lambda cid, nd: cp.impedance(id=cid, nodes=nd, Z=10+3j),
    phase_b=lambda cid, nd: cp.impedance(id=cid, nodes=nd, Z=11+3j),
    phase_c=lambda cid, nd: cp.impedance(id=cid, nodes=nd, Z=12+3j),
)

delta = three_phase_custom_component_delta(
    id="D",
    nodes=("BUS2",),
    phase_a=lambda cid, nd: cp.impedance(id=cid, nodes=nd, Z=20+5j),
    phase_b=lambda cid, nd: cp.impedance(id=cid, nodes=nd, Z=21+5j),
    phase_c=lambda cid, nd: cp.impedance(id=cid, nodes=nd, Z=22+5j),
)

tp = ThreePhaseCircuit(components=[line, star, delta], ground_node="N")
expanded = transform_three_phase_circuit(tp)
expanded
```

```python
for c in expanded.components:
    print(f"{c.id:5s}  {c.nodes}  {c.type}")
```
