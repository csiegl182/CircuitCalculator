# Three-Phase Example (RL load in Y)

```python
import numpy as np

from CircuitCalculator.ThreePhaseCircuit import (
    ThreePhaseCircuit,
    three_phase_voltage_source_star,
    three_phase_impedance_load_star,
    three_phase_complex_solution,
)
```

```python
f = 50
w = 2*np.pi*f
R = 8
L = 0.025
Z_rl = R + 1j*w*L

tp_circuit = ThreePhaseCircuit(
    components=[
        three_phase_voltage_source_star(id="V1", nodes=("BUS", "N"), V=230+0j),
        three_phase_impedance_load_star(id="RL", nodes=("BUS", "N"), Z=Z_rl),
    ],
    ground_node="N",
)

solution = three_phase_complex_solution(tp_circuit, w=w)
```

```python
for phase in ["a", "b", "c"]:
    bid = f"RL.{phase}"
    I = solution.get_current(bid)
    S = solution.get_power(bid)
    print(bid, "I=", I, "S=", S)
```
