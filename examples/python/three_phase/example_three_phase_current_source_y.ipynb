# Three-Phase Example (Y current source + Y impedance load)

```python
import numpy as np

from CircuitCalculator.ThreePhaseCircuit import (
    ThreePhaseCircuit,
    three_phase_current_source_star,
    three_phase_impedance_load_star,
    transform_three_phase_circuit,
    three_phase_complex_solution,
)
```

```python
tp_circuit = ThreePhaseCircuit(
    components=[
        three_phase_current_source_star(id="Iy", nodes=("BUS", "N"), I=8+0j, Y=0.0+0.0j),
        three_phase_impedance_load_star(id="Zy", nodes=("BUS", "N"), Z=12+6j),
    ],
    ground_node="N",
)

expanded = transform_three_phase_circuit(tp_circuit)
print(expanded)
```

```python
solution = three_phase_complex_solution(tp_circuit, w=2*np.pi*50)

for branch in ["Iy.a", "Iy.b", "Iy.c", "Zy.a", "Zy.b", "Zy.c"]:
    print(branch, "V=", solution.get_voltage(branch), "I=", solution.get_current(branch))
```
