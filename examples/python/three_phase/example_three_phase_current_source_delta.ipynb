# Three-Phase Example (Delta current source + Delta voltage source + Delta load)

This setup keeps the delta current source example but adds a delta voltage source, which provides a stable operating point for the current-source branches in the current solver model.

```python
import numpy as np

from CircuitCalculator.ThreePhaseCircuit import (
    ThreePhaseCircuit,
    three_phase_current_source_delta,
    three_phase_voltage_source_delta,
    three_phase_impedance_load_delta,
    transform_three_phase_circuit,
    three_phase_complex_solution,
)
```

```python
tp_circuit = ThreePhaseCircuit(
    components=[
        three_phase_voltage_source_delta(id="Vd", nodes=("BUS",), V=400+0j, Z=0.2+0.05j),
        three_phase_current_source_delta(id="Id", nodes=("BUS",), I=12+0j),
        three_phase_impedance_load_delta(id="Zd", nodes=("BUS",), Z=25+10j),
    ]
)

expanded = transform_three_phase_circuit(tp_circuit)
expanded
```

```python
all_nodes = [n for c in expanded.components for n in c.nodes]
print('total node refs:', len(all_nodes))
print('unique nodes   :', len(set(all_nodes)))
print('nodes          :', sorted(set(all_nodes)))
```

```python
solution = three_phase_complex_solution(tp_circuit, w=2*np.pi*50)

for branch in ["Vd.ab", "Vd.bc", "Vd.ca", "Id.ab", "Id.bc", "Id.ca", "Zd.ab", "Zd.bc", "Zd.ca"]:
    print(branch, "V=", solution.get_voltage(branch), "I=", solution.get_current(branch))
```
