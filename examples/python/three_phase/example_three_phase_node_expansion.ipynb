# Three-Phase Example (Node expansion visibility)

```python
from CircuitCalculator.ThreePhaseCircuit import (
    ThreePhaseCircuit,
    ThreePhaseVoltageSourceY,
    ThreePhaseLoadY,
    transform_three_phase_circuit,
)
```

```python
tp_circuit = ThreePhaseCircuit(
    components=[
        ThreePhaseVoltageSourceY(id="V1", nodes=("MAIN_BUS", "N0"), V=230+0j),
        ThreePhaseLoadY(id="LOAD1", nodes=("MAIN_BUS", "N0"), Z=12+3j),
        ThreePhaseLoadY(id="LOAD2", nodes=("MAIN_BUS", "N0"), Z=20+7j),
    ],
    ground_node="N0",
)

expanded_circuit = transform_three_phase_circuit(tp_circuit)

for c in expanded_circuit.components:
    print(f"{c.id:8s} : {c.nodes}")

print("\nGround node:", expanded_circuit.ground_node)
```
