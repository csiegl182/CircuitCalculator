# Three-Phase Example (Y source + Delta load)

```python
import numpy as np

from CircuitCalculator.ThreePhaseCircuit import (
    ThreePhaseCircuit,
    three_phase_voltage_source_star,
    three_phase_impedance_load_delta,
    transform_three_phase_circuit,
    three_phase_complex_solution,
)
```

```python
tp_circuit = ThreePhaseCircuit(
    components=[
        three_phase_voltage_source_star(id="Vy", nodes=("SRC", "N"), V=230+0j, Z=0.2+0.05j),
        three_phase_impedance_load_delta(id="Zd", nodes=("SRC",), Z=18+6j),
    ],
    ground_node="N",
)

expanded_circuit = transform_three_phase_circuit(tp_circuit)
expanded_circuit
```

```python
solution = three_phase_complex_solution(tp_circuit, w=2*np.pi*50)

for branch in ["Vy.a", "Vy.b", "Vy.c", "Zd.ab", "Zd.bc", "Zd.ca"]:
    print(branch, "I=", solution.get_current(branch))
```
