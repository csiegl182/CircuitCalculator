# Three-Phase Example (Y source + Y load)

```python
import numpy as np

from CircuitCalculator.ThreePhaseCircuit import (
    ThreePhaseCircuit,
    ThreePhaseVoltageSourceY,
    ThreePhaseLoadY,
    transform_three_phase_circuit,
    three_phase_complex_solution,
)
```

```python
tp_circuit = ThreePhaseCircuit(
    components=[
        ThreePhaseVoltageSourceY(id="V1", nodes=("BUS", "N"), V=230+0j, Z=0.2+0.1j),
        ThreePhaseLoadY(id="Z1", nodes=("BUS", "N"), Z=15+4j),
    ],
    ground_node="N",
)

expanded_circuit = transform_three_phase_circuit(tp_circuit)
expanded_circuit
```

```python
solution = three_phase_complex_solution(tp_circuit, w=2*np.pi*50)

for branch in ["V1.a", "V1.b", "V1.c"]:
    print(branch, "I=", solution.get_current(branch))
```
