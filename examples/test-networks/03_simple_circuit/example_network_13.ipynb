```python
from CircuitCalculator.SimpleCircuit.Elements import Schematic, VoltageSource, Resistor, Line, Ground
from CircuitCalculator.SimpleCircuit.DiagramSolution import single_frequency_complex_solution
```

```python
with Schematic(unit=7) as schematic:
    schematic += (U1:=VoltageSource(V=1, R=0, name='U1', reverse=True).up())
    schematic += (R1:=Resistor(R=10, name='R1').right())
    schematic += Line().up().at(U1.end).length(3)
    schematic += (R2:=Resistor(R=20, name='R2').right())
    schematic += Line().down().length(3)
    schematic += (R3:=Resistor(R=30, name='R3').down())
    schematic += Line().left()
    schematic += Ground()
    schemdraw_solution = single_frequency_complex_solution(schematic)
    schematic += schemdraw_solution.draw_voltage('R1')
    schematic += schemdraw_solution.draw_current('R1')
    schematic += schemdraw_solution.draw_voltage('R2')
    schematic += schemdraw_solution.draw_current('R2')
    schematic += schemdraw_solution.draw_voltage('R3')
    schematic += schemdraw_solution.draw_current('R3')
    schematic += schemdraw_solution.draw_voltage('U1')
    schematic += schemdraw_solution.draw_current('U1', reverse=True)
```
