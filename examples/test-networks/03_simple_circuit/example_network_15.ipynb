```python
from CircuitCalculator.SimpleCircuit.styles.styling import select
select(style='DIN')
from CircuitCalculator.SimpleCircuit.Elements import Schematic, VoltageSource, Resistor, Line, Ground, LabeledLine
from CircuitCalculator.SimpleCircuit.DiagramSolution import real_solution
```

```python
with Schematic(unit=5) as schematic:
    schematic += VoltageSource(V=1, name='V1', reverse=True).up(length=10)
    schematic += Line().right()
    schematic += (R1:=Resistor(R=10, name='R1').down())
    schematic += (R2:=Resistor(R=20, name='R2').down())
    schematic += Line().left()
    schematic += Ground()
    schematic += Line().at(R1.start).right()
    schematic += (R3:=Resistor(R=30, name='R3').down())
    schematic += (R4:=Resistor(R=40, name='R4').down())
    schematic += Line().left()
    schematic += LabeledLine(name='bridge').at(R1.end).right()
    schemdraw_solution = real_solution(schematic)
    schematic += schemdraw_solution.draw_voltage('R1')
    schematic += schemdraw_solution.draw_current('R1')
    schematic += schemdraw_solution.draw_voltage('R2')
    schematic += schemdraw_solution.draw_current('R2')
    schematic += schemdraw_solution.draw_voltage('R3')
    schematic += schemdraw_solution.draw_current('R3')
    schematic += schemdraw_solution.draw_voltage('R4')
    schematic += schemdraw_solution.draw_current('R4')
    schematic += schemdraw_solution.draw_current('bridge')
    schematic += schemdraw_solution.draw_current('V1', reverse=True)
```
