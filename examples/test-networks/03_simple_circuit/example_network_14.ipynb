```python
from CircuitCalculator.SimpleCircuit.styles.styling import select
select(style='DIN')
from CircuitCalculator.SimpleCircuit.Elements import Schematic, RealVoltageSource, Resistor, Line, LabeledLine, Ground
from CircuitCalculator.SimpleCircuit.DiagramSolution import single_frequency_complex_solution
```

```python
with Schematic(unit=(unit:=7)) as schematic:
    schematic += (U1:=RealVoltageSource(V=1, name='U1', R=10).up())
    schematic += Line().right()
    schematic += (R1:=Resistor(R=10, name='R1').down())
    schematic += Line().at(R1.start).right()
    schematic += (R2:=Resistor(R=20, name='R2').down())
    schematic += Line().left()
    schematic += Line().left()
    schematic += Ground()
    schematic += LabeledLine(name='L1').up().length(2)
    schemdraw_solution = single_frequency_complex_solution(schematic)
    schematic += schemdraw_solution.draw_voltage('R1')
    schematic += schemdraw_solution.draw_current('R1')
    schematic += schemdraw_solution.draw_voltage('R2')
    schematic += schemdraw_solution.draw_current('R2')
    schematic += schemdraw_solution.draw_voltage('U1', reverse=False)
    schematic += schemdraw_solution.draw_current('L1', reverse=True)
```
