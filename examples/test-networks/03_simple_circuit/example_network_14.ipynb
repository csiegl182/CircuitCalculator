```python
from CircuitCalculator.SimpleCircuit.styles.styling import select
select(style='DIN')
from CircuitCalculator.SimpleCircuit.Elements import Schematic, RealVoltageSource, VoltageSource, Resistor, Line, Ground
from CircuitCalculator.SimpleCircuit.DiagramSolution import single_frequency_solution
```

```python
with Schematic(unit=7) as schematic:
    schematic += (U1:=RealVoltageSource(
        voltage_source=VoltageSource(V=1, name='U1', reverse=True),
        resistor=Resistor(R=10, name='Ri'), reverse=True).up())
    schematic += Line().right()
    schematic += (R1:=Resistor(R=10, name='R1').down())
    schematic += Line().at(R1.start).right()
    schematic += (R2:=Resistor(R=20, name='R2').down())
    schematic += Line().left()
    schematic += Line().left()
    schematic += Ground()
    schemdraw_solution = single_frequency_solution(schematic)
    schematic += schemdraw_solution.draw_voltage('R1')
    schematic += schemdraw_solution.draw_current('R1')
    schematic += schemdraw_solution.draw_voltage('R2')
    schematic += schemdraw_solution.draw_current('R2')
    schematic += schemdraw_solution.draw_voltage('U1', reverse=True)
    schematic += schemdraw_solution.draw_current('U1', reverse=True)
```
