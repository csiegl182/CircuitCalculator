```python
from CircuitCalculator.SimpleCircuit.styles.styling import select
select(style='DIN')
from CircuitCalculator.SimpleCircuit.Elements import Schematic, VoltageSource
from CircuitCalculator.SimpleCircuit.DiagramSolution import real_solution
```

```python
with Schematic(unit=5) as schematic:
    schematic += VoltageSource(V=1, name='Vs', reverse=True).up(length=10)
    schemdraw_solution = real_solution(schematic)
    schematic += schemdraw_solution.draw_voltage('Vs')
    schematic += schemdraw_solution.draw_current('Vs')
    schematic += schemdraw_solution.draw_power('Vs')
```
