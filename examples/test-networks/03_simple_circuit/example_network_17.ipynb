```python
from CircuitCalculator.SimpleCircuit.styles.styling import select
select(style='DIN')
from CircuitCalculator.SimpleCircuit.Elements import Schematic, CurrentSource
from CircuitCalculator.SimpleCircuit.DiagramSolution import real_solution
```

```python
with Schematic(unit=5) as schematic:
    schematic += CurrentSource(I=1, name='Is', reverse=False).up(length=10)
    schemdraw_solution = real_solution(schematic)
    schematic += schemdraw_solution.draw_voltage('Is')
    schematic += schemdraw_solution.draw_current('Is')
    schematic += schemdraw_solution.draw_power('Is')
```
