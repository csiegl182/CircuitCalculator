# Example 26 - Voltage Controlled Current Source

```python
from CircuitCalculator.Circuit.circuit import Circuit
import CircuitCalculator.Circuit.Components.components as ccp
from CircuitCalculator.Circuit.solution import dc_solution
```

```python
circuit = Circuit(
    components=[
        ccp.dc_current_source(id='Vctrl', I=5, nodes=('0', '1')),
        ccp.resistor(id='R1', R=5, nodes=('1', '2')),
        ccp.resistor(id='R2', R=10, nodes=('2', '0')),
        ccp.voltage_controlled_current_source(id='Gm', G=0.5, nodes=('2', '3'), control_nodes=('1', '2')),
        ccp.resistor(id='R3', R=15, nodes=('3', '2'))
    ],
    ground_node='0'
)
```

```python
solution = dc_solution(circuit)
print(f'I(Gm)={solution.get_current("Gm"):2.2f}A')
print(f'V(R1)={solution.get_voltage("R1"):2.2f}V')
print(f'V(R2)={solution.get_voltage("R2"):2.2f}V')
print(f'V(R3)={solution.get_voltage("R3"):2.2f}V')
```
