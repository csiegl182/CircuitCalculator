# Example 30 - Operational Amplifier

```python
from CircuitCalculator.Circuit.circuit import Circuit
import CircuitCalculator.Circuit.Components.components as ccp
from CircuitCalculator.Circuit.solution import dc_solution
```

```python
circuit = Circuit(
    components=[
        ccp.dc_voltage_source(id='Vin', V=1, nodes=('plus', '0')),
        ccp.operational_amplifier(id='U1', nodes=('out', '0'), input_nodes=('plus', 'minus')),
        ccp.resistor(id='Rf', R=9000, nodes=('out', 'minus')),
        ccp.resistor(id='Rg', R=1000, nodes=('minus', '0')),
        ccp.resistor(id='Rload', R=10000, nodes=('out', '0'))
    ],
    ground_node='0'
)
```

```python
solution = dc_solution(circuit)
print(f'V(U1)={solution.get_voltage("U1"):2.2f}V')
print(f'V(Rg)={solution.get_voltage("Rg"):2.2f}V')
print(f'V(diff)={solution.get_potential("plus") - solution.get_potential("minus"):2.6f}V')
print(f'I(U1)={solution.get_current("U1"):2.4f}A')
```
