# Example 24 - Parallel Voltage Sources

```python
from CircuitCalculator.Circuit.circuit import Circuit
import CircuitCalculator.Circuit.Components.components as ccp
from CircuitCalculator.Circuit.solution import dc_solution
```

```python
circuit = Circuit(
    components=[
        ccp.dc_voltage_source(id='Uq1', V=1, nodes=('1', '0')),
        ccp.dc_voltage_source(id='Uq2', V=2, nodes=('1', '0')),
        ccp.resistor(id='R', R=1, nodes=('1', '0')),
    ],
    ground_node='0'
)
```

```python
solution = dc_solution(circuit)
print(solution.get_voltage('Uq1'))
print(solution.get_voltage('Uq2'))
print(solution.get_voltage('R'))
print(solution.get_current('Uq1'))
print(solution.get_current('Uq2'))
print(solution.get_current('R'))
```
