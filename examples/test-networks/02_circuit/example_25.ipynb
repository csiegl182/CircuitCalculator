# Example 25 - Serial Current Sources

```python
from CircuitCalculator.Circuit.circuit import Circuit
import CircuitCalculator.Circuit.Components.components as ccp
from CircuitCalculator.Circuit.solution import dc_solution
```

```python
circuit = Circuit(
    components=[
        ccp.dc_current_source(id='Iq1', I=1, nodes=('1', '0')),
        ccp.dc_current_source(id='Iq2', I=2, nodes=('2', '1')),
        ccp.resistor(id='R', R=1, nodes=('2', '0')),
    ],
    ground_node='0'
)
```

```python
solution = dc_solution(circuit)
print(solution.get_voltage('Iq1'))
print(solution.get_voltage('Iq2'))
print(solution.get_voltage('R'))
print(solution.get_current('Iq1'))
print(solution.get_current('Iq2'))
print(solution.get_current('R'))
```
