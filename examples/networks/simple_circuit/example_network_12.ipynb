```python
from CircuitCalculator.SimpleCircuit.Elements import Schematic, VoltageSource, Line, Ground, Impedance
from CircuitCalculator.SimpleCircuit.DiagramSolution import complex_network_dc_solution
```

```python
with Schematic() as schematic:
    schematic += (U1:=VoltageSource(V=1-2j, name='U1', reverse=True, precision=2).up())
    schematic += Line().right()
    schematic += (Z1:=Impedance(Z=10+10j, name='Z1', precision=2).down())
    schematic += Line().left()
    schematic += Ground()
    schematic += Impedance(Z=20, name='Z2').at(Z1.start).right()
    schematic += Impedance(Z=30, name='Z3').down()
    schematic += Line().left()
    schemdraw_solution = complex_network_dc_solution(schematic)
    schematic += schemdraw_solution.draw_voltage('Z1')
    schematic += schemdraw_solution.draw_current('Z1')
    schematic += schemdraw_solution.draw_voltage('Z2')
    schematic += schemdraw_solution.draw_voltage('Z3')
    schematic += schemdraw_solution.draw_current('Z3', end=True)
    schematic += schemdraw_solution.draw_current('U1')
```
