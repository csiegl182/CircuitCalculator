```python
from CircuitCalculator.SimpleCircuit.styles.styling import select
select(style='DIN')
from CircuitCalculator.SimpleSimulation.simulator import show_schematic_from_file, simulate_schematic_from_file
import matplotlib.pyplot as plt
import numpy as np
```

```python
solution = simulate_schematic_from_file('current_dividor.json', 'dc')
print(f'V(R)={solution.get_voltage("R")}')
```

```python
show_schematic_from_file('rc.yaml')
```

```python
solution = simulate_schematic_from_file('rc.yaml', 'complex', w=500000)
print(f'{solution.get_voltage("R1")}')
```

```python
t = np.linspace(0, 5e-6, 1000)
s = lambda t: np.heaviside(t-1e-6, 1)


solution = simulate_schematic_from_file('rc.yaml', 'transient', tin=t, input={'Uq': s})
v = solution.get_voltage("C1")

plt.plot(t, s(t))
plt.plot(*v)
```

```python
solution = simulate_schematic_from_file('rc_sin.yaml', 'complex', w=500000)
t = np.linspace(0, 5e-6, 1000)

print(f'U(R1)={solution.get_voltage("R1")}')
print(f'U(C1)={solution.get_voltage("C1")}')
```

```python
solution = simulate_schematic_from_file('rc_sin.yaml', 'time_domain', w_max=5000000)
t = np.linspace(0, 50e-6, 1000)

plt.plot(t, solution.get_voltage("R1")(t))
```
