```python
from CircuitCalculator.SimpleCircuit.styles.styling import select
select(style='DIN')
from CircuitCalculator.SimpleCircuit.Elements import Schematic, Line, VoltageSource, Lamp
from CircuitCalculator.SimpleCircuit.LampLighter import light_lamps
```

```python
with Schematic(unit=(unit:=3)) as schematic:
    schematic += VoltageSource(V=10, name='Uq', reverse=True).up(length=2*unit)
    schematic += Line().right()
    schematic += Lamp(P_ref=100, V_ref=10, name='A').down()
    schematic += Lamp(P_ref=100, V_ref=10, name='B').down()
    schematic += Line().left()
    light_lamps(schematic)
```

```python
with Schematic(unit=(unit:=3)) as schematic:
    schematic += VoltageSource(V=10, name='Uq', reverse=True).up(length=4*unit)
    schematic += Line().right(2*unit)
    schematic += (C:=Lamp(P_ref=100, V_ref=10, name='C').down())
    schematic += (D:=Lamp(P_ref=100, V_ref=10, name='D').down())
    schematic += (E:=Lamp(P_ref=100, V_ref=10, name='E').down())
    schematic += (F:=Lamp(P_ref=100, V_ref=10, name='F').down())
    schematic += Line().left(2*unit)
    schematic += Line().left(unit/2).at(C.end)
    schematic += Line().down(2*unit)
    schematic += Line().right(unit/2)
    schematic += Line().right(unit/2).at(D.end)
    schematic += Line().down(2*unit)
    schematic += Line().left(unit/2)
    light_lamps(schematic)
```

```python
with Schematic(unit=(unit:=3)) as schematic:
    schematic += VoltageSource(V=10, name='Uq', reverse=True).up(length=4*unit)
    schematic += Line().right(2*unit)
    schematic += (G:=Lamp(P_ref=100, V_ref=10, name='G').down())
    schematic += (H:=Lamp(P_ref=100, V_ref=10, name='H').down())
    schematic += (I:=Lamp(P_ref=100, V_ref=10, name='I').down())
    schematic += Line().down(unit)
    schematic += Line().left(2*unit)
    schematic += Line().left(unit/2).at(G.end)
    schematic += Line().down(2*unit)
    schematic += Line().right(unit/2)
    schematic += Line().right(unit/2).at(H.end)
    schematic += Line().down(unit)
    schematic += Line().left(unit/2)
    light_lamps(schematic)
```
