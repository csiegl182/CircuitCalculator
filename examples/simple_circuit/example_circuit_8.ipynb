```python
from CircuitCalculator.SimpleCircuit.styles.styling import select
select(style='DIN')
from CircuitCalculator.SimpleCircuit.Elements import Schematic, Resistor, Line, Ground, VoltageSource
from CircuitCalculator.SimpleCircuit.DiagramSolution import single_frequency_complex_solution
from CircuitCalculator.SimpleCircuit.serializer import dump_json, load_json
```

```python
schematic = Schematic(unit=9)
schematic += VoltageSource(V=1, reverse=True, name='Uq')
schematic += Resistor(R=2, name='R1').right()
schematic += Resistor(R=4, name='R2').down()
schematic += Line().left()
schematic += Ground()
dump_json('example_circuit_8.json', schematic, indent=4)
```

```python
schematic2 = load_json('example_circuit_8.json')
schematic_solution = single_frequency_complex_solution(schematic=schematic2)
schematic2 += schematic_solution.draw_voltage('R1')
schematic2 += schematic_solution.draw_voltage('R2')
schematic2 += schematic_solution.draw_voltage('Uq')
schematic2 += schematic_solution.draw_current('R1')
schematic2 += schematic_solution.draw_power('R1')
schematic2 += schematic_solution.draw_power('Uq')
schematic2 += schematic_solution.draw_power('R2')
schematic2.draw()
```
